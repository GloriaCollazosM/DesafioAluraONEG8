### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
sumaVentasTienda = 0
for i in tienda:
  if i == "Precio":
    for j in tienda[i]:
      sumaVentasTienda += j
print('La Venta de la Tienda 1 es: ', sumaVentasTienda)

sumaVentasTienda2 = 0
for i in tienda2:
  if i == "Precio":
    for j in tienda2[i]:
      sumaVentasTienda2 += j
print('La Venta de la Tienda 2 es: ', sumaVentasTienda2)

sumaVentasTienda3 = 0
for i in tienda3:
  if i == "Precio":
    for j in tienda3[i]:
      sumaVentasTienda3 += j
print('La Venta de la Tienda 3 es: ', sumaVentasTienda3)

sumaVentasTienda4 = 0
for i in tienda4:
  if i == "Precio":
    for j in tienda4[i]:
      sumaVentasTienda4 += j
print('La Venta de la Tienda 4 es: ', sumaVentasTienda4)
```

```python
VentasTodas = {sumaVentasTienda:1, sumaVentasTienda2:2, sumaVentasTienda3:3, sumaVentasTienda4:4}
maxima = max(VentasTodas)
if maxima in VentasTodas:
    maxNum = VentasTodas[maxima]
minima = min(VentasTodas)
if minima in VentasTodas:
    minNum = VentasTodas[minima]
print('La tienda con las ventas más altas es la #', maxNum , 'con una venta de $', maxima)
print('La tienda con las ventas más bajas es la #', minNum , 'con una venta de $', minima)
```

# 2. Ventas por categoría

```python
CategTienda = {}
for i in tienda:
  if i == "Categoría del Producto":
    for j in tienda[i]:
      if j in CategTienda:
        CategTienda[j] += 1
      else:
        CategTienda[j] = 1
ordenado = dict(sorted(CategTienda.items(), key=lambda item: item[1], reverse=True))
print('Las ventas más populares por categoría para la Tienda 1 son: ', ordenado)

CategTienda2 = {}
for i in tienda2:
  if i == "Categoría del Producto":
    for j in tienda2[i]:
      if j in CategTienda2:
        CategTienda2[j] += 1
      else:
        CategTienda2[j] = 1
ordenado2 = dict(sorted(CategTienda2.items(), key=lambda item: item[1], reverse=True))
print('Las ventas más populares por categoría para la Tienda 2 son: ', ordenado2)

CategTienda3 = {}
for i in tienda3:
  if i == "Categoría del Producto":
    for j in tienda3[i]:
      if j in CategTienda3:
        CategTienda3[j] += 1
      else:
        CategTienda3[j] = 1
ordenado3 = dict(sorted(CategTienda3.items(), key=lambda item: item[1], reverse=True))
print('Las ventas más populares por categoría para la Tienda 3 son: ', ordenado3)

CategTienda4 = {}
for i in tienda4:
  if i == "Categoría del Producto":
    for j in tienda4[i]:
      if j in CategTienda4:
        CategTienda4[j] += 1
      else:
        CategTienda4[j] = 1
ordenado4 = dict(sorted(CategTienda4.items(), key=lambda item: item[1], reverse=True))
print('Las ventas más populares por categoría para la Tienda 4 son: ', ordenado4)
```

# 3. Calificación promedio de la tienda


```python
def promedio(lista):
    return sum(lista) / len(lista)

CalificacionTienda = []
for i in tienda:
  if i == "Calificación":
    for j in tienda[i]:
      CalificacionTienda.append(j)
prom = round(promedio(CalificacionTienda), 2)
print('La calificación promedio de la Tienda 1 es: ', prom)

CalificacionTienda2 = []
for i in tienda2:
  if i == "Calificación":
    for j in tienda2[i]:
      CalificacionTienda2.append(j)
prom2 = round(promedio(CalificacionTienda2), 2)
print('La calificación promedio de la Tienda 2 es: ', prom2)

CalificacionTienda3 = []
for i in tienda3:
  if i == "Calificación":
    for j in tienda3[i]:
      CalificacionTienda3.append(j)
prom3 = round(promedio(CalificacionTienda3), 2)
print('La calificación promedio de la Tienda 3 es: ', prom3)

CalificacionTienda4 = []
for i in tienda4:
  if i == "Calificación":
    for j in tienda4[i]:
      CalificacionTienda4.append(j)
prom4 = round(promedio(CalificacionTienda4), 2)
print('La calificación promedio de la Tienda 4 es: ', prom4)
```

```python
CalifTodas = {prom:1, prom2:2, prom3:3, prom4:4}
maximaCal = max(CalifTodas)
if maximaCal in CalifTodas:
    maxCalNum = CalifTodas[maximaCal]
minimaCal = min(CalifTodas)
if minimaCal in CalifTodas:
    minCalNum = CalifTodas[minimaCal]
print('La tienda con la calificación más alta es la #', maxCalNum , 'con una calificación de', maximaCal)
print('La tienda con la calificación más baja es la #', minCalNum , 'con una calificación de', minimaCal)
```

# 5. Envío promedio por tienda

```python
def promedio(lista):
    return sum(lista) / len(lista)

EnvioTienda = []
for i in tienda:
  if i == "Costo de envío":
    for j in tienda[i]:
      EnvioTienda.append(j)
promen = round(promedio(EnvioTienda), 2)
print('El costo promedio de los envios en la Tienda 1 es de: $', promen)

EnvioTienda2 = []
for i in tienda2:
  if i == "Costo de envío":
    for j in tienda2[i]:
      EnvioTienda2.append(j)
promen2 = round(promedio(EnvioTienda2), 2)
print('El costo promedio de los envios en la Tienda 2 es de: $', promen2)

EnvioTienda3 = []
for i in tienda3:
  if i == "Costo de envío":
    for j in tienda3[i]:
      EnvioTienda3.append(j)
promen3 = round(promedio(EnvioTienda3), 2)
print('El costo promedio de los envios en la Tienda 3 es de: $', promen3)

EnvioTienda4 = []
for i in tienda4:
  if i == "Costo de envío":
    for j in tienda4[i]:
      EnvioTienda4.append(j)
promen4 = round(promedio(EnvioTienda4), 2)
print('El costo promedio de los envios en la Tienda 4 es de: $', promen4)
```

```python
EnvTodas = {promen:1, promen2:2, promen3:3, promen4:4}
maximaEn = max(EnvTodas)
if maximaEn in EnvTodas:
    maxEnNum = EnvTodas[maximaEn]
minimaEn = min(EnvTodas)
if minimaEn in EnvTodas:
    minEnNum = EnvTodas[minimaEn]
print('La tienda con los envíos más económicos es la #', minEnNum , 'con un promedio de $', minimaEn)
print('La tienda con los envios más costosos es la #', maxEnNum , 'con un promedio de $', maximaEn)
```

## Conclusión y Recomendación

### ¿A qué tienda debe vender el Sr. Juan?

Considerando todos los factores analizados:

- **Tienda 1** tiene los ingresos más altos, pero la calificación promedio más baja y el costo de envío más elevado, lo cual puede afectar la satisfacción y la rentabilidad a largo plazo.  
- **Tienda 3** presenta la mejor calificación de clientes, una venta sólida (tercera en ingresos), y un costo de envío más bajo que las tiendas 1 y 2, además de un buen desempeño en categorías clave.  
- **Tienda 4** ofrece el envío más económico pero cuenta con el menor ingreso total y una calificación promedio moderada.  
- **Tienda 2** está en un punto intermedio en casi todos los factores, sin destacar ni en puntos fuertes ni débiles.

### Recomendación

Se recomienda que el Sr. Juan **venda en la Tienda 3** porque:

- Equilibra ingresos competitivos con la mejor experiencia y satisfacción del cliente.  
- Presenta costos de envío razonables, lo que puede maximizar la rentabilidad.  
- Tiene una sólida presencia en categorías populares como muebles y electrónicos.  
- Los productos más vendidos reflejan una demanda estable y diversa.  

Aunque la Tienda 1 genera más ingresos, la menor calificación y mayores costos logísticos representan riesgos que podrían afectar la sostenibilidad. La Tienda 3 destaca como la opción más balanceada y con mayor potencial de crecimiento sostenido.
